```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import random
import matplotlib.colors as colors
import matplotlib.cm as cmx
#from matplotlib.patches import Polygon
pylab.rcParams['figure.figsize'] = (40., 40.)
```

```python
#Variables:
migration_direction_to_country = True
num_of_arrows = 10
arrow_thickness = 2
a_head_width = 200000
a_head_length = 400000
country_name = 'India' # Use the three letter Code
time = '2000'

font = {'family' : 'Vera Sans',
        'weight' : 'bold',
        'size'   : 22}

matplotlib.rc('font', **font)
```

```python
mig_total =  pd.read_csv('WB-bimigration.csv')

def strip(text):
    try:
        return text.strip()
    except AttributeError:
        return text
cap_info = pd.read_csv('geodata_capitals.csv', sep =";", converters = {'ISO3166A3 ': strip} )

cap_small = cap_info[['ISOen_name','ISO3166A3','BGNc_latitude','BGNc_longitude']]
cap_small = cap_small.rename(columns={'ISOen_name' : 'Country_name', 'ISO3166A3': 'Country_code', 'BGNc_latitude':'lat','BGNc_longitude':'lon'})
cap_small[['Country_code']] = cap_small[['Country_code']].replace(' ', '', regex=True)

mig_total = mig_total.rename(columns={'﻿Country Origin Name': 'Country_orig', 
                                      'Country Origin Code': 'Country_code',
                                      'Country Dest Code': 'Country_d_code',
                                      'All Years' : 'Total',
                                      '1960 [1960]': '1960',
                                      '1970 [1970]': '1970',
                                      '1980 [1980]': '1980',
                                      '1990 [1990]': '1990',
                                      '2000 [2000]': '2000'}) 

mig_sub = mig_total[mig_total['Country_orig']==country_name]
country = mig_sub['Country_code'].iat[0]

if migration_direction_to_country == True:
    mig_to_co = mig_total[mig_total['Country_d_code']==country]
    mig_to_co = mig_to_co[['Country_code', time]]

else: 
    mig_to_co = mig_total[mig_total['Country_code']==country]
    mig_to_co = mig_to_co[['Country_d_code', time]]
    mig_to_co = mig_to_co.rename(columns={'Country_d_code': 'Country_code'})


# this dot in front of the column name posed a problem for a long time
mig_to_co =mig_to_co.reset_index()

mig_to_co = mig_to_co.replace(' ', '', regex=True)

mig_joined = mig_to_co.merge(
    cap_small,
    how='inner',
    on='Country_code')

mig_joined = mig_joined.sort(columns = [time], ascending = False)


mig_joined.head(10)
```

```python
mig_co = mig_joined[mig_joined['Country_code']==country]
mig_co =mig_co.reset_index()


fig = plt.figure() 
axc = fig.add_axes([0.8, 0.1, 0.025, 0.15])
ax = fig.add_axes([0.1, 0.1, 0.7, 0.35])

map = Basemap(projection='moll', lat_0 = 0, lon_0 = 0,
    resolution = 'c',
    llcrnrlon=-180, llcrnrlat=-80.0,
    urcrnrlon=180, urcrnrlat=80, ax =ax)


lon_country = mig_co.loc[0,'lon']
lat_country = mig_co.loc[0,'lat']
x_co, y_co = map(lon_country, lat_country) # this can't before the line where I declare what kind of object map is

map.drawcoastlines()
map.drawcountries()
#map.fillcontinents(color = 'coral')
map.drawmapboundary()

cmap = plt.cm.jet

mig_max_at_time = mig_joined[time].loc[mig_joined[time].idxmax()]
cNorm  = colors.Normalize(vmin=0, vmax=mig_max_at_time) 
# Min and Max of spectrum have to be made dependent on data
scalarMap = cmx.ScalarMappable(norm=cNorm,cmap=cmap)


#Introduce first country
arrows = list(reversed(range(num_of_arrows)))

for pos in arrows:
    colorVal1 =scalarMap.to_rgba(mig_joined[time].iat[pos])
    lon_orig = mig_joined['lon'].iat[pos]
    lat_orig = mig_joined['lat'].iat[pos]
    x_orig,y_orig = map(lon_orig, lat_orig)
    #print x,y,x2,y2
    if migration_direction_to_country == True:
        plt.arrow(x_orig,y_orig,x_co-x_orig,y_co-y_orig, linewidth = arrow_thickness, head_width= a_head_width, head_length =a_head_length, length_includes_head = True, color=colorVal1) # WORKS

    else:
        plt.arrow(x_co,y_co,x_orig-x_co,y_orig-y_co, linewidth = arrow_thickness, head_width=a_head_width, head_length =a_head_length, length_includes_head = True, color=colorVal1) # WORKS

cb1 = mpl.colorbar.ColorbarBase(axc, cmap=cmap, norm=cNorm,orientation='vertical')
if migration_direction_to_country == True:
    title_map = "Migration to %s in the decade up to %s" % (country_name, time)
else:
    title_map = "Migration from %s in the decade up to %s" % (country_name, time)
plt.title(title_map)
title_map_file = title_map + ".png"
plt.savefig(title_map_file)
plt.show()
```
