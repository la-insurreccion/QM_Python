```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import random
import matplotlib.colors as colors
import matplotlib.cm as cmx
```

## Defining the variables of the visualization

The next cell creates variables that determine not only which data is presented in the final visualizations, but also the design and size of it.

```python
# Variables that change the content of the visualization. They affect which data is selected from the database.
migration_direction_to_country = False 
# A True value tells the program that we are interested in Immigration with respect to the country we choose. 
# A False value will change the data selection, so that the Emigration from the specific country is displayed
num_of_arrows = 10
# num_of_arrows tells the computer how many arrows should be displayed. Since we sort the data it will always 
# display the countries which have the highest number of migration with respec to the chosen country in the time period.
country_name = 'Brazil' 
# This variable simply sets the country we want to look at. Displaying all countries would simply be an overload of information.
time = '1960'
# The time variable determines the decade we look at. 

# The following variable do not change the data displayed, but only the the type of the presentation.
# The Size of the arrow, including the head is determined. The font dictionary changes the size of the title and the legend.
arrow_thickness = 2
a_head_width = 200000 # width of the head of the arrows
a_head_length = 400000 # length of the head of the arrows
font = {'family' : 'sans-serif',
        'weight' : 'bold',
        'size'   : 20}

matplotlib.rc('font', **font)
# The following line determines the size of the output figure. 
pylab.rcParams['figure.figsize'] = (18., 9.)
```

## Getting the data into the right form

Now the data is imported from two csv-files. These files are a common format for saving data which is made publicly available. Basically, they are just information seperated by a specific character, e.g. a comma. One csv file contains information about the number of migrants between specific countries and the other about the geographical location of the capitals of all counries.
After they are imported into a dataframe they are cleaned from characters which could produce mistakes further on. The dataframe is a certain way of saving the data which allows for easy manipulation. Also, the names of the columns are changed to make the coding process easier. Finally, they are joined up and sorted to be in the right form for visualization.

```python
# The first line creates a dataframe from a csv file with the name stated in brackets.
mig_total =  pd.read_csv('WB-bimigration.csv')

def strip(text):
    try:
        return text.strip()
    except AttributeError:
        return text
# Here the second csv file is imported as a dataframe. At the same it is stripped from extra white spaces.    
cap_info = pd.read_csv('geodata_capitals.csv', sep =";", converters = {'ISO3166A3 ': strip} )
# A new dataframe is created containing only the necessary columns. The name of the columns is change to prevent spelling problems.
cap_small = cap_info[['ISOen_name','ISO3166A3','BGNc_latitude','BGNc_longitude']]
cap_small = cap_small.rename(columns={'ISOen_name' : 'Country_name', 'ISO3166A3': 'Country_code', 'BGNc_latitude':'lat','BGNc_longitude':'lon'})
cap_small[['Country_code']] = cap_small[['Country_code']].replace(' ', '', regex=True)

mig_total = mig_total.rename(columns={'﻿Country Origin Name': 'Country_orig', 
                                      'Country Origin Code': 'Country_code',
                                      'Country Dest Code': 'Country_d_code',
                                      'All Years' : 'Total',
                                      '1960 [1960]': '1960',
                                      '1970 [1970]': '1970',
                                      '1980 [1980]': '1980',
                                      '1990 [1990]': '1990',
                                      '2000 [2000]': '2000'}) 
# The next lines get the country code of the country stated in the variables cell. 
mig_sub = mig_total[mig_total['Country_orig']==country_name]
country = mig_sub['Country_code'].iat[0]
# Now data is selected from the database depending on whether we want to look at emigration
# or immigration with respect to the chosen country.
if migration_direction_to_country == True:
    mig_to_co = mig_total[mig_total['Country_d_code']==country]
    mig_to_co = mig_to_co[['Country_code', time]]

else: 
    mig_to_co = mig_total[mig_total['Country_code']==country]
    mig_to_co = mig_to_co[['Country_d_code', time]]
    mig_to_co = mig_to_co.rename(columns={'Country_d_code': 'Country_code'})

mig_to_co = mig_to_co.replace(' ', '', regex=True)
# The two dataframes are joined on the basis of the country code, which is a column in both of them. 
mig_joined = mig_to_co.merge(
    cap_small,
    how='inner',
    on='Country_code')
# The resulting dataframe is sorted so that the biggest migration flows can be found at the top.
mig_joined = mig_joined.sort(columns = [time], ascending = False)

# The first ten entries of the dataframe are printed, just to give the user the option to check whether the right data was selected.
mig_joined.head(num_of_arrows)
```

## Creating the visualization

The last cell uses the dataframe created beforehand to draw a map with arrows representing the direction and magnitude of migration from or to the chosen country. This is done by first creating a world map using the mentioned Basemap library. All the information about the country borders are already saved in it. The geographical information are extraced from the dataframe and a loop creates as many arrows to other countries as specified in the variables cell. Finally, a legend is added and the final map is shown in the program and printed to a file. 

```python
# The first lines create an object, which can take two seperate figures, in our case a map and a colorbar as a legend. 
fig = plt.figure() 
axc = fig.add_axes([0.85, 0.1, 0.025, 0.4])
ax = fig.add_axes([0.025, 0.025, 0.80, 0.9])
# The map of the world is created by using a specific projection and being centered on the coordinates 0 North, 0 East.
map = Basemap(projection='moll', lat_0 = 0, lon_0 = 0,
    resolution = 'c',
    llcrnrlon=-180, llcrnrlat=-80.0,
    urcrnrlon=180, urcrnrlat=80, ax =ax)
# The coordinates of the chosen country are extracted from the dataframe.
mig_co = mig_joined[mig_joined['Country_code']==country]
mig_co =mig_co.reset_index()
lon_country = mig_co.loc[0,'lon']
lat_country = mig_co.loc[0,'lat']
x_co, y_co = map(lon_country, lat_country)

# Coastlines and country boundaries are drawn on the map.
map.drawcoastlines()
map.drawcountries()
#map.fillcontinents(color = 'coral')  # A color can be added to the land mass if desired.
map.drawmapboundary()

# Now the colors for the arrows are initiatlised. They are matched to the magnitude of the migration for the specific country. 
cmap = plt.cm.jet
mig_max_at_time = mig_joined[time].loc[mig_joined[time].idxmax()]
cNorm  = colors.Normalize(vmin=100, vmax=mig_max_at_time) 
scalarMap = cmx.ScalarMappable(norm=cNorm,cmap=cmap)

# A simple list is created, which determines the order in which the arrows are drawn. 
# We want the arrows representing the biggest migration flows to be added last,
# so that they are not hidden by less important flows.
arrows = list(reversed(range(num_of_arrows)))
# The for loop extracts the information to draw the migration flows. 
# The number of migrants and the geographical location is needed.
for pos in arrows:
    colorVal1 =scalarMap.to_rgba(mig_joined[time].iat[pos])
    lon_orig = mig_joined['lon'].iat[pos]
    lat_orig = mig_joined['lat'].iat[pos]
    x_orig,y_orig = map(lon_orig, lat_orig)
    # The first if statement increases the thickness for the three most important migration flows by a small number. 
    # This is not explained in the legend. However, it improves the readability of the map. 
    if pos <3:
        arrow_thickness_p = arrow_thickness + 4
    else:
        arrow_thickness_p = arrow_thickness
    # Now the arrows are drawn. The direction is determined on whether emigration or immigration is to be visualized.
    # The arrows are only drawn, if there is a reported migration of at least 100. This makes sure that missing data 
    # does not lead to the appearance of misleading arrows
    if mig_joined[time].iat[pos]>99:
        if migration_direction_to_country == True:
            plt.arrow(x_orig,y_orig,x_co-x_orig,y_co-y_orig, linewidth = arrow_thickness_p, head_width= a_head_width, head_length =a_head_length, length_includes_head = True, color=colorVal1) # WORKS

        else:
            plt.arrow(x_co,y_co,x_orig-x_co,y_orig-y_co, linewidth = arrow_thickness_p, head_width=a_head_width, head_length =a_head_length, length_includes_head = True, color=colorVal1) # WORKS
# The next line creates the colorbar.
cb1 = mpl.colorbar.ColorbarBase(axc, cmap=cmap, norm=cNorm,orientation='vertical')
# The title of the map and the name of the file is created by combining the variables determining the data shown.
num_str = str(num_of_arrows)
if migration_direction_to_country == True:
    title_map = "Top %s Immigration flows to %s in the decade up to ~ %s" % (num_str, country_name, time)
else:
    title_map = "Top %s Emigration flows from %s in the decade up to ~ %s" % (num_str, country_name, time)
plt.title(title_map)
title_map_file = title_map + ".png"
plt.savefig(title_map_file)
plt.show()
```
